1. Create a new DataFrame that only includes customers who:
   - have a **low total_claim_amount** (e.g., below $1,000),
   - have a response "Yes" to the last marketing campaign.

```python
# your code goes here
import pandas as pd
url = "https://raw.githubusercontent.com/data-bootcamp-v4/data/main/marketing_customer_analysis.csv"
customer_df = pd.read_csv(url)
customer_df

pd.options.display.float_format = '{:,.2f}'.format
```

```python
customer_df.shape
```

```python
customer_df.head(5)
```

```python
# Verificamos el nombre de todas las columnas para saber con cuales vamos a trabajar
customer_df.columns
```

```python
# Filtamos las columnas "total_claim_amount" y "response"
filtered_customer_df = customer_df[(customer_df["Total Claim Amount"] < 1000) & (customer_df["Response"] == "Yes")].copy()
```

```python
filtered_customer_df.shape
```

```python
filtered_customer_df.head(5)
```

```python
# Creamos un indice nuevo sin eliminar el anterior.
filtered_customer_df.loc[:, "New Index"] = range(len(filtered_customer_df))

cols =["New Index"] + [col for col in filtered_customer_df.columns if col != "New Index"]
filtered_customer_df = filtered_customer_df[cols]

filtered_customer_df.head(5)
```

```python
# Creamos el dataframe final con el "New_Index" + "Response" + "Total_Clain_Amount"

customer_df_final = filtered_customer_df[["New Index", "Response", "Total Claim Amount"]]

customer_df_final.head(5)
```

2. Using the original Dataframe, analyze:
   - the average `monthly_premium` and/or customer lifetime value by `policy_type` and `gender` for customers who responded "Yes", and
   - compare these insights to `total_claim_amount` patterns, and discuss which segments appear most profitable or low-risk for the company.

```python
# Calculamos los clientes que respondieron SI a la campaña de marketing
customer_yes = customer_df[customer_df["Response"] == "Yes"]
```

```python
# Calculamos el promedio de "monthly_premium" y "customer lifetime value" por "policy_type" y "gender"

average = customer_yes.groupby(["Policy Type", "Gender"])[["Monthly Premium Auto", "Customer Lifetime Value"]].mean().reset_index()

average
```

```python
average_total_claim_amount = customer_yes.groupby(["Policy Type", "Gender"])[["Monthly Premium Auto", "Customer Lifetime Value", "Total Claim Amount"]].mean().reset_index()

average_total_claim_amount
```

En el lado contrario vemos que los cliente con poliza Personal Auto de genero masculino y los clientes de poliza Special Auto de genero femenino representan que son clientes de menor rentabilidad y de un riesgo alto comparado con los demás.

3. Analyze the total number of customers who have policies in each state, and then filter the results to only include states where there are more than 500 customers.

```python
# Analizamos cuantos clientes tiene cada estado
customer_by_state = customer_df.groupby("State")["Customer"].count().reset_index()

customer_by_state.head()
```

```python
# Filtramos los estados que tien mas de 500 clientes
state_over_500_df = customer_by_state[customer_by_state["Customer"] > 500]

state_over_500_df
```

4. Find the maximum, minimum, and median customer lifetime value by education level and gender. Write your conclusions.

```python
value_min_max_median = customer_df.groupby(["Education", "Gender"])[["Customer Lifetime Value"]].agg(["min", "max", "median"]).reset_index()

value_min_max_median
```

En cuanto, al valor max son muy altos en todos los grupos, hay un pico significativo en High School de genero masculino que representa un alto potencial para el negocio aunque podria ser un caso extremo.

5. The marketing team wants to analyze the number of policies sold by state and month. Present the data in a table where the months are arranged as columns and the states are arranged as rows.

```python
sold_policy_type = customer_df.groupby(["State", "Months Since Policy Inception"])["Number of Policies"].count().reset_index()

sold_policy_type.head()
```

```python
# Para que la columna sea mas legible pivotamos la tabla
policy_pivot = sold_policy_type.pivot(index="State", columns="Months Since Policy Inception", values="Number of Policies").fillna(0)

policy_pivot
```

6.  Display a new DataFrame that contains the number of policies sold by month, by state, for the top 3 states with the highest number of policies sold.

*Hint:*
- *To accomplish this, you will first need to group the data by state and month, then count the number of policies sold for each group. Afterwards, you will need to sort the data by the count of policies sold in descending order.*
- *Next, you will select the top 3 states with the highest number of policies sold.*
- *Finally, you will create a new DataFrame that contains the number of policies sold by month for each of the top 3 states.*

```python
sold_policy_type = customer_df.groupby(["State", "Months Since Policy Inception"])["Number of Policies"].count().reset_index()

sold_policy_type
```

```python
total_policies_by_state = sold_policy_type.groupby("State")["Number of Policies"].sum().reset_index()

total_policies_by_state
```

```python
top_states = total_policies_by_state.sort_values(by="Number of Policies", ascending=False).head(3)

top_states
```

```python
top_3_states = top_states["State"].tolist()

policies_top_3 = sold_policy_type[sold_policy_type["State"].isin(top_3_states)]

policies_top_3
```

```python
policies_top_3_pivot = policies_top_3.pivot(index="State", columns="Months Since Policy Inception", values="Number of Policies").fillna(0)

policies_top_3_pivot
```

7. The marketing team wants to analyze the effect of different marketing channels on the customer response rate.

Hint: You can use melt to unpivot the data and create a table that shows the customer response rate (those who responded "Yes") by marketing channel.

```python
# Verificamos las columnas "Sales Channel" y "Response"
customer_df[["Sales Channel", "Response"]].head()
```

```python
# Calculamos la tasa "Yes" por canal
response_yes_rate = (customer_df.groupby("Sales Channel")["Response"].apply(lambda x: (x == "Yes").mean() * 100).reset_index(name="Response Yes Rate (%)"))

print(response_yes_rate)
```
